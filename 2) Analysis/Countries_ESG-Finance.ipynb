```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
```

```python
# Importing the dataset
df = pd.read_csv("../3) Data/main_df_no_NAN_99p.csv").drop(columns="Unnamed: 0")
```

```python
# Summary statistics for each column
df.describe()
```

```python
# Average ESG scores across regions
df_REGION = df.groupby('Region of Headquarters').mean()
esg_cols = ['ESG Score', 'ESG Combined Score', 'ESG Controversies Score']
df_REGION_sorted = df_REGION.sort_values(by='ESG Score', ascending=False)

# Plotting the results
df_REGION_sorted[esg_cols].plot(kind='bar', figsize=(10,6))
plt.title('Average ESG Scores by Region')
plt.xlabel('Region')
plt.ylabel('Score')
plt.show()
```

```python
# Plotting ESG performance vs Total Return for all regions
sns.scatterplot(x='ESG Score', y='Total Return', hue='Region of Headquarters', data=df)
plt.title('ESG Score vs Total Return by Region')
plt.xlabel('ESG Score')
plt.ylabel('Total Return')
plt.show()
```

```python
# Plotting ESG performance vs Total Return for a particular region of interest
Region_of_interest = 'Europe'

sns.scatterplot(x='ESG Score', y='Total Return', hue='Region of Headquarters', data=df[df['Region of Headquarters']==Region_of_interest])
plt.title('ESG Score vs Total Return by Region')
plt.xlabel('ESG Score')
plt.ylabel('Total Return')
plt.show()
```

```python
# Plotting ESG performance vs Revenue for all regions
sns.scatterplot(x='ESG Score', y='Revenue - Mean', hue='Region of Headquarters', data=df)
plt.title('ESG Score vs Revenue by Region')
plt.xlabel('ESG Score')
plt.ylabel('Revenue')
plt.show()
```

```python
# Plotting ESG performance vs Revenue for a region of interest
Region_of_interest = 'Europe'

sns.scatterplot(x='ESG Score', y='Revenue - Mean', hue='Region of Headquarters', data=df[df['Region of Headquarters']==Region_of_interest])
plt.title('ESG Score vs Revenue by Region')
plt.xlabel('ESG Score')
plt.ylabel('Revenue')
plt.show()
```

```python
# Plotting ESG performance vs Return on Equity for all regions
sns.scatterplot(x='ESG Score', y='Return On Equity - Mean', hue='Region of Headquarters', data=df)
plt.title('ESG Score vs Return on Equity by Region')
plt.xlabel('ESG Score')
plt.ylabel('Return On Equity')
plt.show()
```

```python
# Plotting ESG performance vs Market Capitalization for a region of interest
Region_of_interest = 'Europe'

sns.scatterplot(x='ESG Score', y='Company Market Cap', hue='Region of Headquarters', data=df[df['Region of Headquarters']==Region_of_interest])
plt.title('ESG Score vs Market Capitalization by Region')
plt.xlabel('ESG Score')
plt.ylabel('Market Capitalization')
plt.show()
```

```python
# Correlation matrix with both ESG and financial performance indicators [FOR ALL COMPANIES]
cols = ['ESG Score',
        'ESG Combined Score',
        'ESG Controversies Score',
        'Total Return',
        'Revenue - Mean',
        'Earnings Per Share - Mean',
        'Return On Equity - Mean',
        'Volume',
        'Company Market Cap']

df_selected = df[cols]

# Compute correlation matrix
corr_matrix = df_selected.corr()

# Create mask for the upper triangle
mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Plot heatmap
fig, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask, ax=ax, vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix of ESG and Financial Performance Indicators', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Correlation matrix with both ESG and financial performance indicators [FOR TOP-10 MARKET-CAP COMPANIES IN ALL REGIONS]

# Filter top 10 companies by market cap from all regions
top_10 = df.groupby('Region of Headquarters').apply(lambda x: x.nlargest(10, 'Company Market Cap')).reset_index(drop=True)

cols = ['ESG Score',
        'ESG Combined Score',
        'ESG Controversies Score',
        'Total Return',
        'Revenue - Mean',
        'Earnings Per Share - Mean',
        'Return On Equity - Mean',
        'Volume',
        'Company Market Cap']

df_selected = top_10[cols]

corr_matrix = df_selected.corr()

mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask, ax=ax, vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix of ESG and Financial Performance Indicators (Top 10 Companies by Market Cap from All Regions)', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Correlation matrix with both ESG and financial performance indicators [FOR BOTTOM-10 MARKET-CAP COMPANIES IN ALL REGIONS]

# Select bottom 10 companies based on Company Market Cap
bottom_companies = df.groupby('Region of Headquarters')['Company Market Cap'].nsmallest(10)
bottom_company_indices = bottom_companies.index.get_level_values(1)
df_bottom = df.loc[bottom_company_indices]

cols = ['ESG Score',
        'ESG Combined Score',
        'ESG Controversies Score',
        'Total Return',
        'Revenue - Mean',
        'Earnings Per Share - Mean',
        'Return On Equity - Mean',
        'Volume',
        'Company Market Cap']

df_selected = df_bottom[cols]

corr_matrix = df_selected.corr()

mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask, ax=ax, vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix of ESG and Financial Performance Indicators for Bottom 10 Companies by Market Cap', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Correlation matrix with both ESG and financial performance indicators [FOR ALL COMPANIES]
# [FOR A GIVEN REGION OF INTEREST]
region = 'Oceania'

cols = ['ESG Score',
        'ESG Combined Score',
        'ESG Controversies Score',
        'Total Return',
        'Revenue - Mean',
        'Earnings Per Share - Mean',
        'Return On Equity - Mean',
        'Volume',
        'Company Market Cap']

# Filter the data for the given region of headquarters
df_selected = df[df['Region of Headquarters'] == region][cols]

corr_matrix = df_selected.corr()

mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(20, 16))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask, ax=ax, vmin=-1, vmax=1, square=True)
plt.title(f'Correlation Matrix of ESG and Financial Performance Indicators for {region} companies', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Bar chart of revenue mean for the top 10 and bottom 10 companies in region of interest
Region_of_interest = 'Europe'

top_bottom_per_region = df[df['Region of Headquarters']==Region_of_interest].groupby('Region of Headquarters'
                                                                                     ).apply(lambda x: x.nlargest(10, 'Total Return'
                                                                                                                  ).append(x.nsmallest(10, 'Total Return'))
                                                                                                                  )
sns.pairplot(data=top_bottom_per_region, hue='Region of Headquarters', vars=['ESG Score', 'Total Return', 'Earnings Per Share - Mean', 'Return On Equity - Mean'])
plt.show()
```

```python
# Plotting Total return by country and ESG score
df_COUNTRY_ESG = df.groupby(['Country of Headquarters', pd.cut(df['ESG Score'], bins=[0,20,40,60,80,100])]).mean()['Total Return'].unstack()
df_COUNTRY_ESG.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Total Return by Country and ESG Score')
plt.xlabel('Country')
plt.ylabel('Total Return')
plt.legend(title='ESG Score Range')
plt.show()
```

```python
# Plotting ESG scores by country of headquarters (aggregated)
selected_countries = ['India', 'Germany', 'Taiwan', 'Singapore']
esg_by_country = df[df['Country of Headquarters'].isin(selected_countries)]
esg_by_country = esg_by_country.groupby('Country of Headquarters')['ESG Score'].mean().reset_index()
esg_by_country = esg_by_country.sort_values('ESG Score', ascending=False)

plt.figure(figsize=(8, 6))
plt.bar(x=esg_by_country['Country of Headquarters'], height=esg_by_country['ESG Score'])
plt.title('ESG Scores by Country of Headquarters', fontsize=16)
plt.xlabel('Country of Headquarters', fontsize=12)
plt.ylabel('Mean ESG Score', fontsize=12)
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.yticks(fontsize=10)
plt.show()
```

```python
# Plotting average total return by region of headquarters
total_return_by_region = df.groupby('Region of Headquarters')['Total Return'].mean().reset_index()
total_return_by_region = total_return_by_region.sort_values('Total Return', ascending=False)

sns.barplot(data=total_return_by_region, x='Region of Headquarters', y='Total Return', ci=None)
plt.title('Average Total Return by Region of Headquarters')
plt.xlabel('Region of Headquarters')
plt.ylabel('Total Return')
plt.show()
```

```python
# Plotting ESG score distribution by country of headquarters
countries = ['India', 'Germany', 'Taiwan', 'Singapore']
df_countries = df[df['Country of Headquarters'].isin(countries)]

plt.figure(figsize=(12, 6))
sns.histplot(data=df_countries, x='ESG Score', hue='Country of Headquarters', kde=True, multiple='stack')
plt.title('ESG Score Distribution by Country of Headquarters')
plt.xlabel('ESG Score')
plt.ylabel('Frequency')
plt.show()
```

```python

# Group the data by 'Country of Headquarters' and calculate the correlation between 'Total Return' and the three ESG scores
corr_df = df.groupby('Country of Headquarters')[['Total Return', 'ESG Combined Score', 'ESG Score', 'ESG Controversies Score']].corr()

# Extract the correlation coefficients for 'Total Return' and the three ESG scores
corr_total_return = corr_df.loc[(slice(None), 'Total Return'), ['ESG Combined Score', 'ESG Score', 'ESG Controversies Score']]

# Visualize the correlation matrix
fig, ax = plt.subplots(figsize=(8, 30))
sns.heatmap(corr_total_return, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix - Total Return vs. ESG Scores by Country of Headquarters')
plt.xlabel('ESG Scores', fontsize=20)
plt.ylabel('Country of Headquarters')
plt.show()
```
