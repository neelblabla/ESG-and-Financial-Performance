```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the importance values as a dictionary
importance_values = {
    'Governance Pillar Score': 0.12763837497378913,
    'Social Pillar Score': 0.12567001280163312,
    'Social Pillar Score YoY change': 0.12559554512585044,
    'Governance Pillar Score YoY change': 0.12413374368421451,
    'ESG Score YoY change': 0.1107868046196555,
    'ESG Combined Score YoY change': 0.10982956561393757,
    'ESG Score': 0.1094043781077757,
    'ESG Combined Score': 0.10924089371478204,
    'ESG Controversies Score YoY change': 0.036254278921731345,
    'ESG Controversies Score': 0.021446402436630554
}

# Convert the dictionary into a pandas DataFrame
importance_df = pd.DataFrame.from_dict(importance_values, orient='index', columns=['Importance'])

# Sort the values by importance (descending order)
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Reset the index to create a column with the pillar names
importance_df = importance_df.reset_index().rename(columns={'index': 'Pillar'})

# Plot the importance values using seaborn
plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
sns.barplot(x="Importance", y="Pillar", data=importance_df, palette="Blues_d")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.title("Feature Importance")
plt.show()
```
