```python
import pandas as pd
```

```python
df_esg_tr = pd.read_csv("tr_esg_df_cleaned.csv")

df_info = pd.read_csv("comp_info_df.csv")

df_comp_fin = pd.read_csv("comp_financials.csv").rename(columns={"Unnamed: 0": "Year"})
```

```python
df_comb = df_esg_tr.merge(df_comp_fin, on=["key", "Year"], how="left"
).merge(df_info, on=["key"], how="left").drop(columns=["Unnamed: 0_x", "Unnamed: 0_y"])
```

```python
df_comb.head()
```

```python
df_comb.info()
```

```python
df_comb.to_csv("merged_esg_df.csv")
```

```python
# Create a new data frame with only relevant columns to keep
df_relevant = df_comb[['Year',
 'Total Return',
 'ESG Score',
 'ESG Combined Score',
 'ESG Controversies Score',
 'Social Pillar Score',
 'Governance Pillar Score',
 'Environmental Pillar Score',
 'key',
 'Revenue - Mean',
 'Earnings Per Share - Mean',
 'Return On Equity - Mean',
 'Volume',
 'Company Market Cap',
 'EBITDA - Mean',
 'Instrument',
 'Country of Headquarters',
 'Region of Headquarters',
 'TRBC Industry Name',
 'TRBC Industry Group Name',
 'TRBC Business Sector Name',
 'TRBC Economic Sector Name']]

df_relevant.head()
```

```python
df_relevant.to_csv("final_relevant_df.csv")
```
