```python
import pandas as pd
```

```python
df_esg_tr = pd.read_csv("tr_esg_df_cleaned.csv")

df_info = pd.read_csv("comp_info_df.csv")

df_comp_fin = pd.read_csv("comp_financials.csv").rename(columns={"Unnamed: 0": "Year"})
```

```python
df_comb = df_esg_tr.merge(df_comp_fin, on=["key", "Year"], how="left"
).merge(df_info, on=["key"], how="left").drop(columns=["Unnamed: 0_x", "Unnamed: 0_y"])
```
